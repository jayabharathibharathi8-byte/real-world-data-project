Task 5: Statistical Summary

Task 6: Correlation Analysis

```
# Task 6: Correlation Analysis
corr = df[['math score','reading score','writing score']].corr()
corr
```

Task 7: Correlation Heatmap

```
# Task 7: Correlation Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

Task 8: Subject Score Distribution

```
# Task 8: Subject Score Distribution
sns.histplot(df['math score'])
plt.show()
```

Task 9: Gender-wise Performance Analysis

```
# Task 9: Gender-wise Performance Analysis
gender_avg = df.groupby('gender')[['math score','reading score','writing score']].mean()
gender_avg.plot(kind='bar')
plt.show()
```
